```python
#Map() function
List = [1,22,3,42,21,11,5]
#example 1
def Sorted(List):
    List.sort()
    return List

sortedList = list(map(Sorted, [List]))[0]
print(sortedList)


#example 2
def NthTimeList(List):
    return List*5

NthTimeListCreated = list(map(NthTimeList,[List]))[0]
print(NthTimeListCreated)


#example 3
def Square(List):
    return List**2

SquareList = list(map(Square,List))
print(SquareList)



#example 4
def AddNToEachElement(x,n):
    return x + n

AfterAdditionList = list(map(lambda x: AddNToEachElement(x,5),List))
print(AfterAdditionList)


#example 5
def manipulate(x,a,b):
    return x + a - b

manipulation = list(map(lambda x: manipulate(x,5,1),List))
print(manipulation)
```
