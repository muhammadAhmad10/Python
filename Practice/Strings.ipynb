```python
string1 = "Muhammad Ahmad"
string2 = "Learning python"

#concatinating strings
print(string1+" "+string2)

#getting a specific character of string
print(string1[2])

#getting specific characters in reverse/negative order
print(string1[-3])

#getting range of characters from string
print(string2[1:5])

#producing a string multiple times
print(string1*3)

#add new characters in string string
#note -- strings are immuteable mean they cannot be changed.
string2 = string2+' dorn'
print(string2)

#checking whether characters exist in string or not
print('hammad' in string1)

print('Ahmad' not in string1)

#using formatting operators to print strings
print('My name is %s and I am %s'%(string1,string2))

#multiline strings
message = """
My name is Ahmad
I am doing BCS
I am a hostelite
"""
print(message)

#iterating through a string
for letters in string1:
    print(letters)

#comparing two strings
if(string1==string2):
    print('Strings are same')
else:
    print('Strings are not same')


#finding length of string
print(len(string2))
```

```python
string1="Muhammad Ahmad"
string2="Learning python"
#convert string to uppercase
print(string1.upper())

#converting string to lowercase
print(string1.lower())

#partitioning a string
print(string2.partition(' '))

#replacing a part of string 
print(string2.replace('ning','ned'))

#check if string contain numeric characters
print(string1.isnumeric())

#finding a character and returning its index
print(string2.find('g'))

#splitting string from left
print(string1.split(' '))

#checking whether string starts with a specific character
print(string1.startswith('A'))

#removing trailing characters ----?????
print(string1.rstrip())

#finding index of substring
print(string1.index('mmad'))

#print quotes inside a string
print('Muhammad Ahmad studies at \"COMSATS\" Lahore. He is doing \'BCS\' ')

#f-strings 
print(f'{string1} is {string2}')
```
