```python
#set
#set contains unique values only and they are unordered mean indexing does not exist in sets. Sets are immuteable mean their values cannot be changed/updated
Set = {1,2,3,4,5}
print(Set)

#creating empty set
Set1 = set()
print(type(Set1))

#inserting element in set
Set1.add('ahmad')
Set1.add('rashid')
Set1.add('hamza')
Set1.add('rehan')
print(Set1)

#deleting element from set
Set1.remove('rashid')
Set1.discard('rehan')
print(Set1)

#updating element in set
#Sets are immuteable mean their values cannot be changed/updated

#updating python set mean updating set with items from list, tuples into our existing set
List = ['hamid','shihid']
Set1.update(List)
print(Set1)

#finding largest item in set
print(max(Set1))

#finding smallest item in set
print(min(Set1))

#checking if any element of set is true. return false only if all elements are false or set is empty
print(any(Set1))

#checking whether all items of set are true or not. return true only if all elements are true or set is empty
Set2 = {'ahmad',False}
print(all(Set1))
print(all(Set2))

#getting length of list
print(len(Set1))

#sorting the set items
print(sorted(Set1))

#returning sum of all items of set and it works only if all items are numeric
Set3 = {1,3,6,7.3}
print(sum(Set3))

#converting a set to list using enumerate() method
list1 = enumerate(Set1)
print(list(list1))


#iterating over a set
for x in Set1:
    print(x)

S1 = {'ahmad','rashid'}
S2 = {'rehan','ahmad'}

#union of sets or combining two sets
print(S1.union(S2))

#intersection of sets or finding/combining only same/common elements
print(S1.intersection(S2))

#difference between sets
print(S1.difference(S2))

#finding symmetric difference or finding/combining elements of two sets without including common element
print(S1.symmetric_difference(S2))

#checking if two sets are equal or not
if(S1==S2):
    print('sets are equal')
else:
    print('sets are not equal')
```
