```python

import pandas as pd         #importing pandas to get data from files
titanicData  = pd.read_csv('E:/Python/Data files/titanic.csv')      #reading data from local path from file titcanic.csv

#getting different statistics from titanic passenger data file (titanic.csv). 
# Note - these statistical methods can only be applied to numneric data only

titanicData['age'].mean()       #getting the average age of titanic passengers

#getting the average age and fare of titanic passengers. Note - if accessing multiple columns then write those 
# column names inside [] brackets inside main [] brackets like titanicData[['age','fare']]
titanicData[['age','fare']].median()        #getting median of age and fare of passenger data

titanicData[['age','fare']].describe()      #getting statistical data of age and fare columns.

#Instead of the predefined statistics, specific combinations of 
#aggregating statistics for given columns can be defined using the DataFrame.agg() method:
titanicData.agg(
    {
        'age': ['min','max','std'],
        'fare': ['mean','count','skew']
    }
)

#use of groupby() function
titanicData[['age','sex']].groupby('sex').mean()    #getting the the mean age of passenger based on gender using groupby() property
titanicData.groupby('sex').mean()       #getting the mean of every column in passenger data based on sex of passengers. Note - it will be applied only on numberic values
#above functionality can also be used as below to exclude data types other than numberic data
titanicData.groupby('sex').mean(numeric_only = True)
titanicData.groupby('sex')['age'].mean()        #getting average age only based on sex

titanicData.groupby(['sex','pclass'])['fare'].mean()        #getting the mean of fare of passengers based on sex and pclass collectively. we need to provide column list to groupby() in this case

titanicData['pclass'].value_counts()        #getting the number of passengers in each pclass
titanicData['sex'].value_counts()           #getting the number of passengers on based of sex
titanicData.groupby('sex').size()           #getting the number of passengers based on sex using groupby() and size() collectively

```
