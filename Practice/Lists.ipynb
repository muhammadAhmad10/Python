```python
List = ['ahmad','rashid']
List1 = ['rehan','hamid']

#defining a list
print(List)

#iterating through a list
for x in List:
    print(x)

#accessing elements of list
print(List[1])

#accessing elements of list in reverse/negative order
print(List[-2])

#defining an empty list
List2 = list()
print(type(List2))

#inserting elements in list
List2.append('abdullah')
List2.append('mahed')
List2.append('hassan')
List2.append('taimoor')
print(List2)

#accessing a section of list 
print(List2[1:3])

#combining two lists
List1.extend(List2)
print(List1)

#updating list items
List2[1] = 'masood'
print(List2)

#deleting item from list
del List2[1]
print(List2)
List2.remove('taimoor')
print(List2)

#removing item from specific index
List1.pop(1)
print(List1)

#clearing/ removing all item from list
print(List2.clear())

#sorting the list items
List1.sort()
print(List1)

#reverse the list items
List1.reverse()
print(List1)

#return the copy of list items
print(List1.copy())

#count the number of specific item from the list mean how many time that item exits
L1 = [1,2,4,2,5,3,3,1,2]
print(L1.count(2))

#finding the index of item of list
print(List1.index('hassan'))
```
