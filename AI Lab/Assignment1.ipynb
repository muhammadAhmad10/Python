```python
graph = {
    'Oradea':{'Zerind':71,'Sibiu':151},
    'Zerind':{'Arad':75,'Oradea':71},
    'Arad':{'Timisoara':118,'Sibiu':140,'Zerind':75},
    'Sibiu':{'Fagaraw':99,'Rimnicu Vilcea':80,'Oradea':151,'Arad':140},
    'Timisoara':{'Lugoj':111,'Arad':118},
    'Lugoj':{'Timisoara':111,'Mehadia':70},
    'Rimnicu Vilcea':{'Sibiu':80,'Craiova':146,'Pitesti':97},
    'Mehadia':{'Lugoj':70,'Drobeta':75},
    'Drobeta':{'Mehadia':75,'Craiova':120},
    'Pitesti':{'Rimnicu Vilcea':97,'Craiova':138,'Bucharest':101},
    'Craiova':{'Drobeta':120,'Rimnicu Vilcea':146,'Pitesti':138},
    'Fagaraw':{'Sibiu':99,'Bucharest':211},
    'Bucharest':{'Pitesti':101,'Giurgin':90,'Fagaraw':211,'Urziceni':85},
    'Giurgin':{'Bucharest':90},
    'Urziceni':{'Bucharest':85,'Hirsova':98,'Vaslui':142},
    'Eforie':{'Hirsova':86},
    'Hirsova':{'Eforie':86,'Urziceni':98},
    'Vaslui':{'Urziceni':142,'Iasi':92},
    'Iasi':{'Vaslui':92,'Neamt':87},
    'Neamt':{'Iasi':87}
}

# initial = 'Arad'
# goat = 'Bucharest'
# def q1():
#     visited = set()
#     visited.add(initial)
#     for x in graph:
#         for y in graph[x]:
#             if x in graph[x][y]:




def cost():
    totalCost = 0
    for x in graph:
        for y in graph[x]:
            totalCost=totalCost+graph[x][y]
    print(totalCost)

cost()



initialState = 'Arad'
goadState = 'Bucharest'

def DFS(graph, initialState, goalState, visited=set()):
    visited.add(initialState)
    if initialState == goalState:
        return visited
    for neighbour in graph[initialState]:
        if neighbour not in visited:
            DFS(graph, neighbour, goalState, visited)
    return visited


def BFS(graph, initialState, goalState):
    explored = set()
    queue = [initialState]

    while queue:
        node = queue.pop(0)
        if node not in explored:
            explored.add(node)
            neighbours = graph[node]
            for neighbour in neighbours:
                queue.append(neighbour)
                if neighbour == goalState:
                    explored.add(neighbour)
                    return explored

    return explored


print('Path with BFS is:', BFS(graph, initialState=initialState,goalState=goadState))
print('Path with DFS is:', DFS(graph, initialState=initialState,goalState=goadState))
```
